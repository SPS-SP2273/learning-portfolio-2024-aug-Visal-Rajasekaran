```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='.', markerfacecolor='yellow',
             markersize = 15,
             ecolor='black',capsize=5,
             label='Y values with error bars')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
x = np.arange(0,50, 0.05)
y = (np.exp(-x/10)) * (np.sin(x))

plt.plot(x,y)
plt.title("$y = e^{-x/10}sin(x)$", fontsize = 20)
plt.xlabel('x', fontsize = 15, color = (1,0.5,0))
plt.ylabel('y', fontsize = 15, rotation = 0, color = (1,0.5,0))
plt.ylim(-1.0,1.0)
plt.grid(alpha = 0.25)
plt.show()
```
