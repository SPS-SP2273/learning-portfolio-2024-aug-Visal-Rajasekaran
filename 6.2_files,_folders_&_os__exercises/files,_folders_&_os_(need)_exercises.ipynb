$\textbf{TASK 2}$

```python
#Preferred method for file renaming using function
def rename_my_file(old_file_name):
    
    split_file_name = old_file_name.split(sep='-')
    
    if len(split_file_name) == 3 and split_file_name[0].isnumeric() : #Rough check of whether file has already been renamed
        return old_file_name
        
    
    month = split_file_name[0]
    date, country_and_ext = split_file_name[1].split(sep='_')
    new_file_name = month_dict[month] + '-' + date + '-' + country_and_ext
    return new_file_name
    
    
```

$\textbf{TASK 4}$

```python
#Note to self. This below is the recommended style of renaming.

file_names = glob.glob('*.txt')

for old_file_name in file_names:
    try:
        new_file_name = rename_my_file(old_file_name)
        if not os.path.exists(new_file_name):     #Seems redundant but may save some time. Skips file if it is already renamed
            os.rename(old_file_name, new_file_name)
    except Exception as err:
        print("Error ",err)
    
```
