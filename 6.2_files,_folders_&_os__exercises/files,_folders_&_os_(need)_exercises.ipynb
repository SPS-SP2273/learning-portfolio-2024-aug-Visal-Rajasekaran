```python
def rename_my_file(old_file_name):
    
    split_file_name = old_file_name.split(sep='-')
    
    if len(split_file_name) == 3 and split_file_name[0].isnumeric() : #Rough check of whether file has been renamed
        return old_file_name
        
    
    month = split_file_name[0]
    date, country_and_ext = split_file_name[1].split(sep='_')
    new_file_name = month_dict[month] + '-' + date + '-' + country_and_ext
    return new_file_name
    
    
```
